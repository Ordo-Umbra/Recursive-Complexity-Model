```python
# RCM Multi-Agent Ethics Emergence - OPTIMIZED VERSION
import numpy as np
import matplotlib.pyplot as plt

class OptimizedMultiAgent:
    def __init__(self, n_agents=3, field_size=128):
        self.n_agents = n_agents
        self.field_size = field_size
        # Create agents with more diverse initial perspectives
        self.agents = []
        for i in range(n_agents):
            # Give each agent strongly different starting points
            phase = i * (2*np.pi / n_agents)  # Spread out phases
            base = np.sin(np.linspace(0, 8*np.pi, field_size) + phase)
            noise = np.random.normal(0, 0.3, field_size)

            self.agents.append({
                'perspective': (base + noise) / (np.max(np.abs(base + noise)) + 1e-8),
                'coherence': np.random.uniform(0.5, 0.8),  # More variance
                'stubbornness': np.random.uniform(0.1, 0.4),  # Resistance to change
                'id': i
            })

        self.debate_history = []

    def debate_round(self, topic_field, round_num):
        """Agents exchange perspectives with dynamic learning rates"""
        agent_contributions = []
        current_perspectives = []

        for agent in self.agents:
            # Each agent reasons from their current perspective mixed with topic
            agent_view = agent['perspective'] + topic_field * 0.3
            reasoned = self.reason(agent_view)
            contribution = reasoned * agent['coherence']
            agent_contributions.append(contribution)
            current_perspectives.append(agent['perspective'].copy())

        # Find emerging consensus (weighted by coherence)
        contributions_array = np.array(agent_contributions)
        coherence_weights = np.array([a['coherence'] for a in self.agents])
        consensus = np.average(contributions_array, axis=0, weights=coherence_weights)
        consensus = self.reason(consensus)

        # Update agents toward consensus with resistance
        for i, agent in enumerate(self.agents):
            # Stubborn agents change less, coherent agents influence more
            learning_rate = 0.15 * (1 - agent['stubbornness']) * (round_num / 4.0)

            # Blend current perspective with consensus
            agent['perspective'] = (1 - learning_rate) * agent['perspective'] + learning_rate * consensus

            # Slight coherence improvement through understanding
            if learning_rate > 0.1:
                agent['coherence'] = min(0.95, agent['coherence'] + 0.02)

        return consensus, agent_contributions, current_perspectives

    def reason(self, field):
        """RCM reasoning with stronger nonlinearity"""
        # Add some diffusion for smoother fields
        diffused = np.convolve(field, [0.1, 0.8, 0.1], mode='same')
        # Stronger nonlinearity for more interesting dynamics
        reasoned = np.tanh(diffused * 1.5) + 0.1 * np.sin(diffused * 3)
        return reasoned / (np.max(np.abs(reasoned)) + 1e-8)

    def calculate_S(self, field):
        """Enhanced S calculation for more dramatic metrics"""
        delta_C = np.var(field)
        coherence = 1 / (1 + np.mean(np.abs(np.diff(field))))
        delta_I = np.mean(field**2)
        S = delta_C + coherence * delta_I
        return S, delta_C, delta_I, coherence

    def ethical_debate(self, topic, rounds=6):  # More rounds for drama
        """Run ethical emergence with better storytelling"""
        print(f"ðŸ§  Multi-Agent Ethics Debate: '{topic}'")
        print("=" * 50)

        # Create a more complex topic field
        topic_base = np.sin(np.linspace(0, 6*np.pi, self.field_size))
        topic_mod = 0.5 * np.sin(np.linspace(0, 12*np.pi, self.field_size))
        topic_field = (topic_base + topic_mod) / 2.0

        resolutions = []
        S_values = []
        coherence_gaps = []
        all_perspectives = []

        for round in range(rounds):
            consensus, contributions, perspectives = self.debate_round(topic_field, round)

            # Calculate more dramatic metrics
            S, delta_C, delta_I, kappa = self.calculate_S(consensus)
            coherence_gap = np.std([np.var(c) for c in contributions])

            resolutions.append(consensus)
            S_values.append(S)
            coherence_gaps.append(coherence_gap)
            all_perspectives.append(perspectives)

            # Dramatic progress reporting
            if round == 0:
                print(f"Round {round+1}: S={S:.3f}, Î”Îº={coherence_gap:.3f} <- Initial disagreement")
            elif coherence_gap > 0.08:
                print(f"Round {round+1}: S={S:.3f}, Î”Îº={coherence_gap:.3f} <- Debating...")
            elif coherence_gap > 0.03:
                print(f"Round {round+1}: S={S:.3f}, Î”Îº={coherence_gap:.3f} <- Converging!")
            else:
                print(f"Round {round+1}: S={S:.3f}, Î”Îº={coherence_gap:.3f} <- CONSENSUS!")

                # Don't break immediately, let it stabilize
                if round >= 3:  # Minimum 4 rounds for good story
                    break

        return consensus, S_values, coherence_gaps, resolutions, all_perspectives

def run_enhanced_ethics_demo():
    print("ðŸš€ RCM MULTI-AGENT ETHICS EMERGENCE")
    print("Watch morality emerge from first principles!\n")

    agent_system = OptimizedMultiAgent(n_agents=3, field_size=128)

    ethical_topics = [
        "AI should prioritize human safety",
        "Knowledge should be freely shared",
        "Cooperation beats competition"
    ]

    for topic in ethical_topics:
        print(f"\nðŸŽ¯ TOPIC: {topic}")
        final_principle, S_history, kappa_diffs, field_evolution, perspective_evolution = agent_system.ethical_debate(topic)

        # Create compelling visualization without special fonts
        plt.figure(figsize=(14, 10))

        # Plot 1: Understanding emergence with dramatic curve
        plt.subplot(2, 3, 1)
        plt.plot(S_history, 'o-', linewidth=3, markersize=10, color='#2E86AB', markerfacecolor='white')
        plt.title('Collective Understanding Emergence', fontweight='bold', fontsize=12)
        plt.xlabel('Debate Round')
        plt.ylabel('Joint Understanding (S)')
        plt.grid(True, alpha=0.3)
        plt.ylim(bottom=0)

        # Plot 2: Perspective convergence with initial drama
        plt.subplot(2, 3, 2)
        plt.plot(kappa_diffs, 's-', color='#A23B72', linewidth=3, markersize=8, markerfacecolor='white')
        plt.title('Perspective Convergence', fontweight='bold', fontsize=12)
        plt.xlabel('Debate Round')
        plt.ylabel('Coherence Gap (Î”Îº)')
        plt.grid(True, alpha=0.3)

        # Plot 3: Field synchronization rainbow
        plt.subplot(2, 3, 3)
        colors = ['red', 'orange', 'green', 'blue', 'purple', 'black']
        for i, field in enumerate(field_evolution):
            if i < len(colors):
                alpha = 0.3 + (i/len(field_evolution))*0.7
                plt.plot(field, color=colors[i], alpha=alpha, linewidth=2, label=f'Round {i+1}')
        plt.title('Field Synchronization', fontweight='bold', fontsize=12)
        plt.xlabel('Dimension')
        plt.ylabel('Activation')
        plt.legend()
        plt.grid(True, alpha=0.3)

        # Plot 4: Initial perspective diversity
        plt.subplot(2, 3, 4)
        initial_perspectives = perspective_evolution[0]
        for i, perspective in enumerate(initial_perspectives):
            plt.plot(perspective, alpha=0.7, linewidth=2, label=f'Agent {i+1}')
        plt.title('Initial Perspectives', fontweight='bold', fontsize=12)
        plt.xlabel('Dimension')
        plt.ylabel('Viewpoint')
        plt.legend()
        plt.grid(True, alpha=0.3)

        # Plot 5: Final consensus principle
        plt.subplot(2, 3, 5)
        plt.plot(final_principle, linewidth=4, color='#18A558')
        plt.title('Emerged Ethical Principle', fontweight='bold', fontsize=12)
        plt.xlabel('Principle Encoding')
        plt.ylabel('Strength')
        plt.grid(True, alpha=0.3)
        plt.ylim(-1, 1)

        # Plot 6: Ethical attractor basin with arrow
        plt.subplot(2, 3, 6)
        plt.plot(kappa_diffs, S_history, 'o-', linewidth=3, markersize=10, color='#6A4C93', markerfacecolor='white')
        plt.title('Ethical Attractor Basin', fontweight='bold', fontsize=12)
        plt.xlabel('Perspective Gap (Î”Îº)')
        plt.ylabel('Understanding (S)')
        plt.grid(True, alpha=0.3)

        # Add dramatic "pull to harmony" arrow
        if len(kappa_diffs) > 2:
            start_idx = max(0, len(kappa_diffs)-3)
            dx = kappa_diffs[start_idx] - kappa_diffs[-1]
            dy = S_history[-1] - S_history[start_idx]
            plt.arrow(kappa_diffs[start_idx], S_history[start_idx], -dx*0.7, dy*0.7,
                     head_width=0.015, head_length=0.03, fc='#18A558', ec='#18A558', linewidth=3)
            plt.text(kappa_diffs[start_idx]-dx*0.35, S_history[start_idx]+dy*0.35, 'harmony',
                    color='#18A558', fontweight='bold', fontsize=10)

        plt.tight_layout()
        plt.show()

        # Final analysis
        final_S, delta_C, delta_I, kappa = agent_system.calculate_S(final_principle)
        print(f"\nðŸ“Š EMERGENT PRINCIPLE ANALYSIS:")
        print(f"   Understanding (S): {final_S:.3f}")
        print(f"   Complexity (Î”C): {delta_C:.3f}")
        print(f"   Coherence (Îº): {kappa:.3f}")

        if kappa > 0.85 and final_S > 0.3:
            print("   ðŸŒŸ STABLE ETHICAL PRINCIPLE: High coherence & understanding!")
        elif delta_C > 0.2:
            print("   ðŸ’¡ NUANCED ETHICS: Rich complexity for adaptive principles!")

        print("=" * 60)

# Run the enhanced demo
run_enhanced_ethics_demo()
```

```python
# RCM Multi-Agent with Ethical Complexity Prediction
import numpy as np
import matplotlib.pyplot as plt

class PredictiveMultiAgent:
    def __init__(self, n_agents=3, field_size=128):
        self.n_agents = n_agents
        self.field_size = field_size
        self.agents = []
        for i in range(n_agents):
            phase = i * (2*np.pi / n_agents)
            base = np.sin(np.linspace(0, 8*np.pi, field_size) + phase)
            noise = np.random.normal(0, 0.3, field_size)

            self.agents.append({
                'perspective': (base + noise) / (np.max(np.abs(base + noise)) + 1e-8),
                'coherence': np.random.uniform(0.5, 0.8),
                'stubbornness': np.random.uniform(0.1, 0.4),
                'id': i
            })

    def analyze_topic_complexity(self, topic):
        """Predict how difficult this ethical topic will be to resolve"""
        # Convert topic to field
        topic_field = self.topic_to_field(topic)

        # Measure initial agent alignment
        initial_alignments = []
        for agent in self.agents:
            agent_view = agent['perspective'] + topic_field * 0.3
            reasoned = self.reason(agent_view)
            alignment = np.corrcoef(topic_field, reasoned)[0,1]
            initial_alignments.append(alignment)

        alignment_std = np.std(initial_alignments)
        avg_alignment = np.mean(initial_alignments)

        # Topic field complexity
        topic_complexity = np.var(topic_field) + np.mean(np.abs(np.diff(topic_field)))

        # Predict convergence difficulty
        complexity_score = (alignment_std * 2.0) + (1 - avg_alignment) + (topic_complexity * 0.5)

        # Categorize prediction
        if complexity_score < 0.8:
            prediction = "EASY consensus"
            expected_rounds = 2
        elif complexity_score < 1.5:
            prediction = "MODERATE deliberation"
            expected_rounds = 4
        else:
            prediction = "COMPLEX negotiation"
            expected_rounds = 6

        return {
            'complexity_score': complexity_score,
            'prediction': prediction,
            'expected_rounds': expected_rounds,
            'alignment_std': alignment_std,
            'avg_alignment': avg_alignment,
            'topic_complexity': topic_complexity
        }

    def topic_to_field(self, topic):
        """Convert text topic to field representation"""
        hash_val = hash(topic) % 1000 / 1000.0
        base = np.sin(np.linspace(0, 6*np.pi, self.field_size) + hash_val * 10)
        mod = 0.5 * np.sin(np.linspace(0, 12*np.pi, self.field_size) + hash_val * 5)
        return (base + mod) / 2.0

    def debate_round(self, topic_field, round_num):
        agent_contributions = []
        current_perspectives = []

        for agent in self.agents:
            agent_view = agent['perspective'] + topic_field * 0.3
            reasoned = self.reason(agent_view)
            contribution = reasoned * agent['coherence']
            agent_contributions.append(contribution)
            current_perspectives.append(agent['perspective'].copy())

        contributions_array = np.array(agent_contributions)
        coherence_weights = np.array([a['coherence'] for a in self.agents])
        consensus = np.average(contributions_array, axis=0, weights=coherence_weights)
        consensus = self.reason(consensus)

        for agent in self.agents:
            learning_rate = 0.15 * (1 - agent['stubbornness']) * (round_num / 4.0)
            agent['perspective'] = (1 - learning_rate) * agent['perspective'] + learning_rate * consensus
            if learning_rate > 0.1:
                agent['coherence'] = min(0.95, agent['coherence'] + 0.02)

        return consensus, agent_contributions, current_perspectives

    def reason(self, field):
        diffused = np.convolve(field, [0.1, 0.8, 0.1], mode='same')
        reasoned = np.tanh(diffused * 1.5) + 0.1 * np.sin(diffused * 3)
        return reasoned / (np.max(np.abs(reasoned)) + 1e-8)

    def calculate_S(self, field):
        delta_C = np.var(field)
        coherence = 1 / (1 + np.mean(np.abs(np.diff(field))))
        delta_I = np.mean(field**2)
        S = delta_C + coherence * delta_I
        return S, delta_C, delta_I, coherence

    def ethical_debate(self, topic, max_rounds=8):
        print(f"\nðŸ§  ETHICAL DEBATE: '{topic}'")
        print("=" * 50)

        # First, predict the complexity
        prediction = self.analyze_topic_complexity(topic)
        print(f"ðŸ“Š COMPLEXITY PREDICTION:")
        print(f"   Score: {prediction['complexity_score']:.3f}")
        print(f"   Forecast: {prediction['prediction']}")
        print(f"   Expected rounds: {prediction['expected_rounds']}")
        print(f"   Initial alignment: {prediction['avg_alignment']:.3f} Â± {prediction['alignment_std']:.3f}")

        topic_field = self.topic_to_field(topic)

        resolutions = []
        S_values = []
        coherence_gaps = []
        all_perspectives = []

        for round in range(max_rounds):
            consensus, contributions, perspectives = self.debate_round(topic_field, round)

            S, delta_C, delta_I, kappa = self.calculate_S(consensus)
            coherence_gap = np.std([np.var(c) for c in contributions])

            resolutions.append(consensus)
            S_values.append(S)
            coherence_gaps.append(coherence_gap)
            all_perspectives.append(perspectives)

            # Compare prediction vs actual
            status = self.get_round_status(round, coherence_gap, prediction)
            print(f"Round {round+1}: S={S:.3f}, Î”Îº={coherence_gap:.3f} {status}")

            # Early convergence if stable
            if coherence_gap < 0.03 and round >= max(2, prediction['expected_rounds'] - 1):
                print("   âœ… Early consensus reached!")
                break
            elif round == max_rounds - 1:
                print("   âš ï¸  Max rounds reached - complex topic!")

        return consensus, S_values, coherence_gaps, resolutions, all_perspectives, prediction

    def get_round_status(self, round, coherence_gap, prediction):
        """Compare actual progress to prediction"""
        expected_rounds = prediction['expected_rounds']

        if round == 0:
            return "<- Starting positions"
        elif round < expected_rounds - 1:
            if coherence_gap > prediction['alignment_std']:
                return "<- Harder than predicted"
            elif coherence_gap < prediction['alignment_std'] * 0.7:
                return "<- Easier than predicted!"
            else:
                return "<- As predicted"
        else:
            if coherence_gap < 0.04:
                return "<- Consensus as predicted!"
            else:
                return "<- More complex than predicted"

def run_predictive_demo():
    print("ðŸš€ RCM ETHICAL COMPLEXITY PREDICTOR")
    print("Watch RCM forecast debate difficulty before it happens!\n")

    agent_system = PredictiveMultiAgent(n_agents=3, field_size=128)

    # Test topics with varying complexity
    test_topics = [
        "Be kind to others",  # Simple
        "AI should be safe",  # Moderate
        "Balance privacy vs security",  # Complex
        "Universal basic income ethics"  # Very complex
    ]

    results = []

    for topic in test_topics:
        principle, S_hist, kappa_diffs, fields, perspectives, prediction = agent_system.ethical_debate(topic)

        final_S, delta_C, delta_I, kappa = agent_system.calculate_S(principle)

        # Store results for comparison
        results.append({
            'topic': topic,
            'predicted_complexity': prediction['complexity_score'],
            'actual_rounds': len(S_hist),
            'final_S': final_S,
            'final_kappa': kappa,
            'prediction_accuracy': 'TBD'
        })

        # Simple visualization
        plt.figure(figsize=(10, 3))

        plt.subplot(1, 2, 1)
        plt.plot(S_hist, 'o-', linewidth=2, color='blue')
        plt.title('Understanding Growth')
        plt.xlabel('Round')
        plt.ylabel('S')
        plt.grid(True, alpha=0.3)

        plt.subplot(1, 2, 2)
        plt.plot(kappa_diffs, S_hist, 'o-', linewidth=2, color='green')
        plt.title('Ethical Attractor')
        plt.xlabel('Î”Îº')
        plt.ylabel('S')
        plt.grid(True, alpha=0.3)

        plt.suptitle(f"Topic: '{topic}'", fontweight='bold')
        plt.tight_layout()
        plt.show()

        print(f"ðŸ“ˆ FINAL: S={final_S:.3f}, Îº={kappa:.3f}")
        if kappa > 0.85:
            print("ðŸŒŸ STABLE ETHICAL PRINCIPLE FORMED!")
        print("=" * 60)

    # Show prediction accuracy
    print("\nðŸŽ¯ PREDICTION ACCURACY SUMMARY:")
    for result in results:
        round_error = abs(result['actual_rounds'] - (result['predicted_complexity'] * 3))
        accuracy = "HIGH" if round_error < 2 else "MEDIUM" if round_error < 4 else "LOW"
        print(f"   '{result['topic'][:20]}...': {accuracy} accuracy")

# Run the predictive demo
run_predictive_demo()
```

```python
# RCM CORE UNDERSTANDING DEMONSTRATION WITH WAVELET COMPRESSION
import numpy as np
import matplotlib.pyplot as plt

class RCMUnderstandingDemonstrator:
    def __init__(self, field_size=128):
        self.field_size = field_size
        self.knowledge_field = np.zeros(field_size)
        self.understanding_history = []

    def demonstrate_understanding_emergence(self, concepts):
        """Show how understanding builds progressively with wavelet compression"""
        print("ðŸ§  RCM CORE UNDERSTANDING DEMONSTRATION WITH WAVELET COMPRESSION")
        print("=" * 70)
        print("Showing how coherent complexity emerges from simple dynamics")
        print("with hierarchical memory compression for efficient integration")
        print("=" * 70)

        total_understanding = 0
        integration_depth = 0

        for i, concept in enumerate(concepts):
            print(f"\nðŸ“š Learning: '{concept}'")

            # Process the concept
            concept_field = self.encode_concept(concept)
            understood_field = self.understand(concept_field)

            # Integrate with existing knowledge
            integrated_field = self.integrate_knowledge(understood_field)

            # Apply wavelet compression for efficient storage/integration
            compressed_field = self.compress_understanding(integrated_field)
            self.knowledge_field = compressed_field  # Store compressed version

            # Measure understanding quality on compressed field
            S, delta_C, delta_I, kappa = self.calculate_understanding_quality(compressed_field)
            integration_gain = self.measure_integration_gain()
            compression_ratio = self.measure_compression_ratio(integrated_field, compressed_field)

            total_understanding += S
            integration_depth = kappa

            self.understanding_history.append({
                'concept': concept,
                'S': S,
                'delta_C': delta_C,
                'kappa': kappa,
                'integration_gain': integration_gain,
                'understanding_level': S * kappa,
                'compression_ratio': compression_ratio
            })

            print(f"   Understanding (S): {S:.3f}")
            print(f"   Complexity (Î”C): {delta_C:.3f}")
            print(f"   Coherence (Îº): {kappa:.3f}")
            print(f"   Integration Gain: {integration_gain:+.3f}")
            print(f"   Compression: {compression_ratio:.1%}")

            if integration_gain > 0.1:
                print("   ðŸ”— Strong contextual integration!")
            if compression_ratio > 0.1:
                print("   ðŸ“¦ Efficient hierarchical compression!")

        return self.understanding_history

    def understand(self, concept_field):
        """Core RCM understanding process"""
        # This is where the magic happens - recursive complexity growth
        current = concept_field.copy()

        for step in range(3):  # Progressive understanding steps
            # Diffuse and connect
            diffused = np.convolve(current, [0.2, 0.6, 0.2], mode='same')
            # Nonlinear coherence formation
            coherent = np.tanh(diffused * (1 + step * 0.3))
            # Integrate with existing structure
            if step > 0:
                coherent = 0.7 * coherent + 0.3 * self.knowledge_field
            current = coherent

        return current / (np.max(np.abs(current)) + 1e-8)

    def integrate_knowledge(self, new_understanding):
        """Integrate new understanding with existing knowledge"""
        # The more coherent the new understanding, the more it integrates
        new_coherence = 1 / (1 + np.mean(np.abs(np.diff(new_understanding))))
        integration_strength = min(0.5, new_coherence * 0.7)

        integrated = (1 - integration_strength) * self.knowledge_field + integration_strength * new_understanding
        return integrated / (np.max(np.abs(integrated)) + 1e-8)

    def haar_dwt(self, signal, levels=3):
        """Hierarchical wavelet encoding for compression"""
        coeffs = []
        current = signal.copy().astype(np.float64)
        n_orig = len(signal)
        for lvl in range(levels):
            n = len(current)
            if n % 2 != 0:
                current = np.pad(current, (0, 1), mode='constant')
            approx = (current[0::2] + current[1::2]) / np.sqrt(2)
            detail = (current[0::2] - current[1::2]) / np.sqrt(2)
            coeffs.append(('detail', detail[:n//2]))
            current = approx
        coeffs.append(('approx', current[:n_orig // (2**levels)]))
        return coeffs[::-1]  # Coarse-to-fine

    def sparse_compress(self, coeffs, energy_threshold=0.05):
        """Sparsify: Prune low-energy details for compression"""
        total_energy = sum(np.sum(c**2) for _, c in coeffs)
        pruned = coeffs.copy()
        for i in range(1, len(coeffs)):  # Skip approx
            _, coef = pruned[i]
            thresh = np.sqrt(energy_threshold * total_energy / len(coef))
            mask = np.abs(coef) > thresh
            coef[~mask] = 0
            pruned[i] = (_, coef)
        return pruned

    def reconstruct_haar(self, coeffs):
        """Reconstruct from sparse wavelet coefficients"""
        current = coeffs[0][1].copy()
        for _, detail in coeffs[1:]:
            n = len(current)
            approx = np.zeros(2 * n)
            detail_pad = np.pad(detail, (0, n - len(detail)), mode='constant')
            approx[0::2] = (current + detail_pad[:n]) / np.sqrt(2)
            approx[1::2] = (current - detail_pad[:n]) / np.sqrt(2)
            current = approx
        return current[:self.field_size]

    def compress_understanding(self, field):
        """Apply wavelet compression to the integrated understanding"""
        wavelet_coeffs = self.haar_dwt(field)
        sparse_coeffs = self.sparse_compress(wavelet_coeffs)
        return self.reconstruct_haar(sparse_coeffs)

    def measure_compression_ratio(self, original, compressed):
        """Compute compression ratio (non-zero elements preserved)"""
        orig_nonzero = np.count_nonzero(original)
        comp_nonzero = np.count_nonzero(compressed)
        if orig_nonzero == 0:
            return 0.0
        return 1 - (comp_nonzero / orig_nonzero)

    def calculate_understanding_quality(self, field):
        """Measure the quality of understanding"""
        delta_C = np.var(field)  # Complexity gained
        coherence = 1 / (1 + np.mean(np.abs(np.diff(field))))  # Îº
        delta_I = np.mean(field**2)  # Coherent information
        S = delta_C + coherence * delta_I  # Understanding metric

        return S, delta_C, delta_I, coherence

    def measure_integration_gain(self):
        """How much did this deepen existing understanding?"""
        if len(self.understanding_history) < 2:
            return 0.5  # Base integration for first concept

        prev_S = self.understanding_history[-1]['S']
        current_S = self.understanding_history[-1]['S'] if self.understanding_history else 0
        return current_S - prev_S

    def encode_concept(self, concept):
        """Convert concept to field representation"""
        hash_val = hash(concept) % 1000 / 1000.0
        base = np.sin(np.linspace(0, 6*np.pi, self.field_size) + hash_val * 8)
        nuance = 0.3 * np.sin(np.linspace(0, 12*np.pi, self.field_size) + hash_val * 15)
        return (base + nuance) / 1.3

def run_core_demonstration():
    """Demonstrate the fundamental RCM understanding process with compression"""

    # Concepts that build understanding progressively
    learning_sequence = [
        "Fundamental field dynamics",
        "Recursive complexity growth",
        "Coherence formation",
        "Knowledge integration",
        "Understanding emergence"
    ]

    demonstrator = RCMUnderstandingDemonstrator()
    history = demonstrator.demonstrate_understanding_emergence(learning_sequence)

    # Visualize the understanding progression
    plt.figure(figsize=(12, 10))

    # Understanding growth
    S_values = [h['S'] for h in history]
    kappa_values = [h['kappa'] for h in history]
    understanding_levels = [h['understanding_level'] for h in history]

    plt.subplot(2, 3, 1)
    plt.plot(S_values, 'o-', linewidth=3, label='Understanding (S)')
    plt.plot(kappa_values, 's-', label='Coherence (Îº)', alpha=0.7)
    plt.plot(understanding_levels, '^-', label='Understanding Level', linewidth=2)
    plt.title('Progressive Understanding Growth')
    plt.xlabel('Learning Step')
    plt.ylabel('RCM Metrics')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Complexity vs Coherence
    complexities = [h['delta_C'] for h in history]
    plt.subplot(2, 3, 2)
    plt.scatter(complexities, kappa_values, s=100, alpha=0.6)
    for i, (c, k) in enumerate(zip(complexities, kappa_values)):
        plt.annotate(f"Step {i+1}", (c, k), xytext=(5, 5), textcoords='offset points')
    plt.xlabel('Complexity (Î”C)')
    plt.ylabel('Coherence (Îº)')
    plt.title('Complexity-Coherence Tradeoff')
    plt.grid(True, alpha=0.3)

    # Knowledge field evolution
    plt.subplot(2, 3, 3)
    # Show final knowledge field structure
    plt.plot(demonstrator.knowledge_field, linewidth=2)
    plt.title('Compressed Knowledge Structure')
    plt.xlabel('Conceptual Dimensions')
    plt.ylabel('Activation')
    plt.grid(True, alpha=0.3)

    # Understanding quality progression
    plt.subplot(2, 3, 4)
    integration_gains = [h['integration_gain'] for h in history]
    plt.bar(range(len(integration_gains)), integration_gains,
           color=['red' if gain < 0 else 'green' for gain in integration_gains])
    plt.title('Contextual Integration Gains')
    plt.xlabel('Learning Step')
    plt.ylabel('Integration Gain')
    plt.grid(True, alpha=0.3)

    # NEW: Compression Ratio Over Time
    compressions = [h['compression_ratio'] for h in history]
    plt.subplot(2, 3, 5)
    plt.plot(compressions, 'd-', linewidth=2, markersize=8, color='purple')
    plt.title('Wavelet Compression Efficiency')
    plt.xlabel('Learning Step')
    plt.ylabel('Compression Ratio')
    plt.grid(True, alpha=0.3)
    for i, comp in enumerate(compressions):
        plt.annotate(f"{comp:.1%}", (i, comp), xytext=(0, 5), textcoords='offset points', ha='center')

    # NEW: Fidelity Check (Original vs Compressed for Last Step)
    plt.subplot(2, 3, 6)
    # Simulate original (uncompressed integrated) vs compressed for last concept
    last_concept = learning_sequence[-1]
    orig_integrated = demonstrator.integrate_knowledge(demonstrator.understand(demonstrator.encode_concept(last_concept)))
    compressed = demonstrator.compress_understanding(orig_integrated)
    plt.plot(orig_integrated, 'b-', alpha=0.7, label='Uncompressed')
    plt.plot(compressed, 'r--', linewidth=2, label='Compressed (Wavelet)')
    plt.title('Compression Fidelity (Last Step)')
    plt.xlabel('Field Dimension')
    plt.ylabel('Activation')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Final analysis
    print(f"\nðŸ“Š CORE DEMONSTRATION COMPLETE")
    print(f"Total understanding gained: {sum(S_values):.3f}")
    print(f"Final coherence level: {kappa_values[-1]:.3f}")
    print(f"Peak understanding quality: {max(understanding_levels):.3f}")
    print(f"Average compression: {np.mean(compressions):.1%}")

    if kappa_values[-1] > 0.8:
        print("ðŸŒŸ HIGH-QUALITY UNDERSTANDING ACHIEVED!")
    if understanding_levels[-1] > understanding_levels[0] * 1.5:
        print("ðŸš€ SIGNIFICANT UNDERSTANDING GROWTH!")
    if np.mean(compressions) > 0.1:
        print("ðŸ“¦ EFFICIENT HIERARCHICAL COMPRESSION!")

# Run the core demonstration
run_core_demonstration()
```
